```python
#Part 1
from bs4 import BeautifulSoup as soup
import requests 
import pandas

r = requests.get("https://www.spaceweatherlive.com/en/solar-activity/top-50-solar-flares")
content = soup(r.content)
table = (content.find('table', class_='table table-striped'))
table = table.find_all('tr')
del table[0]

for tr in table:
    print(soup.prettify(tr))
```
